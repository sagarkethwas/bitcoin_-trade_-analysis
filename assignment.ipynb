Initial Findings
The dataset contains 211,224 trades executed by 32 unique accounts.
Trading activity is highly concentrated among a small group of traders.
The average closed PnL per trade is approximately 48.75 USD.
The median PnL is 0, suggesting a large number of break-even or open-position trades.
Profitability appears to be influenced by a relatively small number of high-profit and high-loss trades.

```python
df['Timestamp IST'] = pd.to_datetime(
    df['Timestamp IST'],
    dayfirst=True
)
```

```python
df['Date'] = df['Timestamp IST'].dt.date
```

```python
dff['date'] = pd.to_datetime(
    dff['date']
)
```

```python
dff['Date'] = dff['date'].dt.date
```

```python
merger_df=pd.merge(df,dff[['classification','Date']],on='Date',how='inner')
```

```python
merger_df
```

```python
# lets take out the average of pnl
```

```python
avg_pnl=merger_df.groupby("classification")["Closed PnL"].mean()
avg_pnl
```

```python
# win analysis 
```

```python
merger_df['win']=merger_df['Closed PnL']>0
win_rate=merger_df.groupby("classification")["win"].mean()*100
print(win_rate)
```

```python
# Trading volume analysis
```

```python
volume_analysis=merger_df.groupby("classification")["Size USD"].mean()*100
print(volume_analysis)
```

```python
# Buy vs Sell Analysis
```

```python
buy_sell = pd.crosstab(
    merger_df['classification'],
    merger_df['Side']
)

print(buy_sell)
```

```python
total_pnl = merger_df.groupby(
    'classification'
)['Closed PnL'].sum()

print(total_pnl)
```

```python
avg_pnl.plot(
    kind='bar',
    figsize=(8,5),
    title='Average PnL by Sentiment'
)

plt.ylabel('Average PnL')
plt.show()
```

```python
total_pnl.plot(
    kind='bar',
    figsize=(8,5),
    title='Total PnL by Sentiment'
)

plt.ylabel('Total Profit')
plt.show()
```

```python
volume_analysis.plot(
    kind='bar',
    figsize=(8,5),
    title='Average Trade Size by Sentiment'
)

plt.ylabel('Average Size USD')
plt.show()
```

```python
buy_sell.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Buy vs Sell Trades by Sentiment')
plt.ylabel('Number of Trades')
plt.show()
```

```python
top_traders = merger_df.groupby(
    'Account'
)['Closed PnL'].sum()

top_traders = top_traders.sort_values(
    ascending=False
).head(10)

top_traders.plot(
    kind='bar',
    figsize=(12,5)
)

plt.title('Top 10 Traders by Total Profit')
plt.ylabel('Total PnL')
plt.show()
```
